```python
import pandas as pd

# Load the CSV file (ensure the file is in the same directory or update the path)
df = pd.read_csv('IndianFoodDatasetCSV.csv')

# Define a keyword-to-base-price mapping in rupees.
price_mapping = {
    'dosa': 40,
    'idli': 20,
    'biryani': 80,
    'pasta': 60,
    'pizza': 70,
    'chole': 50,
    'sambar': 30,
    'rice': 30,
    'wheat': 25,
    'dal': 20,
    'sandwich': 40,
    'salad': 50,
    'cake': 50,
    'snack': 15,
    'curd': 20,
    'noodles': 35,
    'curry': 30,
    'grill': 40,
    'burger': 45,
    'roll': 35,
    'paneer': 40,
    'momos': 30
    # Expand with more keywords as needed.
}

def get_base_price(product_name):
    """Determine base price (in rupees) based on product name using keyword matching."""
    name = str(product_name).lower()  # Ensure matching is case-insensitive
    for keyword, base_price in price_mapping.items():
        if keyword in name:
            return base_price
    # Default base price if no keywords match
    return 30

def suggest_resell_price(row):
    """Compute the Average_Resell_Price using base price, perishability, and time of day adjustments."""
    base_price = get_base_price(row['Product_Name'])

    # Adjust price based on Perishability.
    # Expected values: 'high', 'medium', 'low'
    perishability = str(row.get('Perishability', 'medium')).lower()
    if perishability == 'high':
        price = base_price  # No discount for highly perishable items
    elif perishability == 'medium':
        price = base_price * 0.9  # Small discount
    elif perishability == 'low':
        price = base_price * 0.8  # Larger discount for items that last longer
    else:
        price = base_price  # Fallback in case of unexpected value

    # Further adjust based on Time_of_Day.
    # Expected values: 'breakfast', 'lunch', 'dinner', or 'any'
    time_of_day = str(row.get('Time_of_Day', 'any')).lower()
    if time_of_day == 'breakfast':
        price += 1
    elif time_of_day == 'lunch':
        price += 2
    elif time_of_day == 'dinner':
        price += 3
    # No adjustment if time_of_day is 'any' or not recognized

    return round(price, 2)

# Apply the pricing function to each row in the CSV.
df['Average_Resell_Price'] = df.apply(suggest_resell_price, axis=1)

# Save the updated DataFrame to a new CSV file.
df.to_csv('IndianFoodDatasetCSV_with_prices.csv', index=False)

print("CSV updated with the Average_Resell_Price column (prices in rupees).")
```

```python
import pandas as pd
from fuzzywuzzy import process

# Load your CSV data
df = pd.read_csv('/content/IndianFoodDatasetCSV_with_prices.csv')  # Replace with the path to your CSV file

# Function to get the closest match and price suggestion
def suggest_price(food_item):
    # Get a list of food names from the database
    food_names = df['TranslatedRecipeName'].tolist()

    # Use fuzzywuzzy to get the best match for the input food item
    match, score = process.extractOne(food_item, food_names)

    # If the match score is above a threshold (e.g., 80), we consider it a good match
    if score >= 80:
        # Find the row that corresponds to the match
        matched_row = df[df['TranslatedRecipeName'] == match]

        # Get the average resell price for that item
        suggested_price = matched_row['Average_Resell_Price'].values[0]
        return match, suggested_price
    else:
        return "No good match found", None

# Example: User inputs a food item
user_input = "noodles egg"  # Example user input (note the typo)
match, price = suggest_price(user_input)

if price is not None:
    print(f"Suggested Price for '{match}': {price}")
else:
    print(match)
```
